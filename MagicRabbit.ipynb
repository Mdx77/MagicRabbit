# Magic Rabbit

### Game created by N. Courvoisier, 2019
### (with basis from code demo at http://www.raywenderlich.com)

```python
# ---------------------------------------
# Part needed to create single .exe file
import sys
import os

if getattr(sys, 'frozen', False):
    os.chdir(sys._MEIPASS)
   
def resource_path(relative):
    if hasattr(sys, "_MEIPASS"):
        return os.path.join(sys._MEIPASS, relative)
    return os.path.join(relative)
# ---------------------------------------

# Import library
import pygame
from pygame.locals import *
import math
import random

# Set Colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
BLUE = (50, 50, 255)
RED = (255, 0, 0)
GREEN = (0, 255, 0)

# Initialize the game
pygame.init()
width, height = 800, 600 # 640, 480
screen=pygame.display.set_mode((width, height))

# Set variables
keys = [False, False, False, False]
playerpos=[150,300]

acc=[0,0]
arrows=[]

badtimer=300
badtimer1=0
badguys=[[800,100]] 
healthvalue=194
hit_counter=20

medtimer=2500 #25sec
meds=[[600,300]]

cartimer=3000 #30sec
cars=[[900,600]] # not visible initially
carpause=300
carhitcounter=0

cloudtimer=3500 #35sec
clouds=[[50,-412]]
mult=1
directioncounter=0    

pygame.mixer.init()
time_game=90000
myClock = pygame.time.Clock()

# Load images
data_dir_images='data/images'
player = pygame.image.load(resource_path(os.path.join(data_dir_images, 'dude.png')))
player2 = pygame.image.load(resource_path(os.path.join(data_dir_images, 'dude2.png')))
grass = pygame.image.load(resource_path(os.path.join(data_dir_images, 'grass2.png')))
castle = pygame.image.load(resource_path(os.path.join(data_dir_images, 'castle.png')))
arrow = pygame.image.load(resource_path(os.path.join(data_dir_images, 'bullet.png')))
carimg1 = pygame.image.load(resource_path(os.path.join(data_dir_images, 'car.png')))
carimg=carimg1
badguyimg1 = pygame.image.load(resource_path(os.path.join(data_dir_images, 'badguy.png')))
badguyimg=badguyimg1
healthbar = pygame.image.load(resource_path(os.path.join(data_dir_images, 'healthbar.png')))
health = pygame.image.load(resource_path(os.path.join(data_dir_images, 'health.png')))
pointer = pygame.image.load(resource_path(os.path.join(data_dir_images, 'pointer.png')))
cloudimg = pygame.image.load(resource_path(os.path.join(data_dir_images, 'cloud.png')))
medimg1 = pygame.image.load(resource_path(os.path.join(data_dir_images, 'med.png')))
medimg=medimg1
gameover = pygame.image.load(resource_path(os.path.join(data_dir_images, 'gameover.png')))
youwin = pygame.image.load(resource_path(os.path.join(data_dir_images, 'youwin.png')))

# Load audio
data_dir_audio='data/audio'
hit = pygame.mixer.Sound(resource_path(os.path.join(data_dir_audio, 'explode.wav')))
enemy = pygame.mixer.Sound(resource_path(os.path.join(data_dir_audio, 'enemy.wav')))
shoot = pygame.mixer.Sound(resource_path(os.path.join(data_dir_audio, 'shoot.wav')))
scream = pygame.mixer.Sound(resource_path(os.path.join(data_dir_audio, 'scream.wav')))
scream6 = pygame.mixer.Sound(resource_path(os.path.join(data_dir_audio, 'scream6.wav')))
jump = pygame.mixer.Sound(resource_path(os.path.join(data_dir_audio, 'jump.wav')))
popout = pygame.mixer.Sound(resource_path(os.path.join(data_dir_audio, 'popout.wav')))
carstart = pygame.mixer.Sound(resource_path(os.path.join(data_dir_audio, 'carstart.wav')))
trumpets = pygame.mixer.Sound(resource_path(os.path.join(data_dir_audio, 'trumpets.wav')))
emptychamber = pygame.mixer.Sound(resource_path(os.path.join(data_dir_audio, 'emptychamber.wav')))

moonlight = resource_path(os.path.join(data_dir_audio, 'moonlight.wav')) # music
pygame.mixer.music.load(moonlight)
pygame.mixer.music.play(-1, 0.0)
pygame.mixer.music.set_volume(0.20)

hit.set_volume(0.05)
enemy.set_volume(0.05)
shoot.set_volume(0.05)
scream.set_volume(0.15)
scream6.set_volume(0.20)
jump.set_volume(0.20)
popout.set_volume(0.15)
carstart.set_volume(0.15)
trumpets.set_volume(0.15)
emptychamber.set_volume(0.15)

# Mouse
pygame.mouse.set_visible( False ) #makes mouse invisible
pointerImg_rect = pointer.get_rect() # mouse

# Instructions
# --------------
pygame.display.set_caption("Magic Rabbit - Instructions")
data_dir_other='data/other'
FreeSansBold=resource_path(os.path.join(data_dir_other, 'FreeSansBold.ttf'))
font_tiny = pygame.font.Font(FreeSansBold, 16)
font_verysmall = pygame.font.Font(FreeSansBold, 24)
font_small = pygame.font.Font(FreeSansBold, 36)
font_big = pygame.font.Font(FreeSansBold, 50)
display_instructions = True
i=0
while display_instructions:
    for event in pygame.event.get():
        if event.type == pygame.KEYDOWN:
            if event.key==K_SPACE:
                display_instructions = False
            elif event.key==K_ESCAPE:
                display_instructions = False

    screen.fill(BLACK)
    text = font_big.render("MAGIC RABBIT", True, WHITE)
    screen.blit(text, [10, 10])
    text = font_verysmall.render("Shoot the badgers coming before they hit your bases.", True, WHITE)
    screen.blit(text, [10, 60])
    text = font_verysmall.render("Credits:", True, WHITE)
    screen.blit(text, [10, 90])
    text = font_tiny.render("- a shot costs 1 credit", True, WHITE)
    screen.blit(text, [30, 120])
    text = font_tiny.render("- a badger killed brings 5 credits, a carrot eaten 20 credits", True, WHITE)
    screen.blit(text, [30, 140])
    text = font_tiny.render("- 20 credits at start", True, WHITE)
    screen.blit(text, [30, 160])
    text = font_verysmall.render("Commands:", True, WHITE)
    screen.blit(text, [10, 190])
    text = font_tiny.render("- WASD or keyboard arrows to move", True, WHITE)
    screen.blit(text, [30, 220])
    text = font_tiny.render("- mouse to change shooting direction, left-click to fire", True, WHITE)
    screen.blit(text, [30, 240])
    text = font_tiny.render("- right-click to atomize everything, watch out the raised ears when ready ;)", True, WHITE)
    screen.blit(text, [30, 260])
    text = font_verysmall.render("Health:", True, WHITE)
    screen.blit(text, [10, 290])
    text = font_tiny.render("- health loss if a bagger reaches a castle", True, WHITE)
    screen.blit(text, [30, 320])
    text = font_tiny.render("- big health loss in case of collision", True, WHITE)
    screen.blit(text, [30, 340])
    text = font_tiny.render("- game over if health is entirely lost", True, WHITE)
    screen.blit(text, [30, 360])
    text = font_tiny.render("- a carrot give half health back and some credits, but sensible to shots!", True, WHITE)
    screen.blit(text, [30, 380])
    text = font_tiny.render("- don't let cars run you over, they block arrows, destroy them for extra carrots...", True, WHITE)
    screen.blit(text, [30, 400])
    text = font_verysmall.render("Shooting accuracy provided at the end for own ranking contest.", True, WHITE)
    screen.blit(text, [10, 430])
    text = font_verysmall.render("Done by Nicolas Courvoisier, 2019", True, WHITE)
    screen.blit(text, [10, 480])
    text = font_tiny.render("(with basis from code demo at http://www.raywenderlich.com)", True, WHITE)
    screen.blit(text, [10, 510])
    if i>=400:
        i=0
    else:
        i+=1
    text_points=i//100*'.'
    text = font_verysmall.render("Press SPACE to start now"+text_points, True, WHITE)
    screen.blit(text, [10, 560])
    pygame.display.flip()
# --------------

# Main loop
# -------------- start of loop
pygame.display.set_caption("Magic Rabbit - Play")
running = 1
exitcode = 0
start_ticks=pygame.time.get_ticks()
while running:
    badtimer-=1
    medtimer-=1
    cartimer-=1
    cloudtimer-=1
    
    # Clear the screen before drawing it again
    screen.fill(0)
    
    # Draw part
    # --------------
    # Draw the screen elements
    grass_width=grass.get_width()
    grass_height=grass.get_height()
    for x in range(width//grass_width+1):
        for y in range(height//grass_height+1):
            screen.blit(grass,(x*grass_width,y*grass_height))
    screen.blit(castle,(0,30))
    screen.blit(castle,(0,145))
    screen.blit(castle,(0,260))
    screen.blit(castle,(0,375))
    screen.blit(castle,(0,490))
    
    # Mouse
    pointerImg_rect.center = pygame.mouse.get_pos()
    screen.blit(pointer, pointerImg_rect)
    
    # Set player position and rotation
    position = pygame.mouse.get_pos()
    angle = math.atan2(position[1]-(playerpos[1]+32),position[0]-(playerpos[0]+26))
    
    if hit_counter>=100:
        playerchoice=player2
    else:
        playerchoice=player
    playerrot = pygame.transform.rotate(playerchoice, 360-angle*57.29)
    playerpos1 = (playerpos[0]-playerrot.get_rect().width/2, playerpos[1]-playerrot.get_rect().height/2)
    screen.blit(playerrot, playerpos1) 

    # Draw arrows
    for bullet in arrows:
        index=0
        velx=math.cos(bullet[0])*10
        vely=math.sin(bullet[0])*10
        bullet[1]+=velx
        bullet[2]+=vely
        if bullet[1]<-64 or bullet[1]>800 or bullet[2]<-64 or bullet[2]>600:
            arrows.pop(index)
        index+=1
        for projectile in arrows:
            arrow1 = pygame.transform.rotate(arrow, 360-projectile[0]*57.29)
            screen.blit(arrow1, (projectile[1], projectile[2]))
    
    # Draw carrots
    if medtimer==0:
        meds.append([random.randint(200,725), random.randint(75,525)])
        jump.play()
        medtimer=1200 #12sec
        
    # Draw car
    if cartimer==80:
        carstart.play() #warning in advance it's coming
    if cartimer==0:
        cars.clear() # remove all other cars if any
        cars.append([random.randint(200,700), 600])
        carpause=300
        cartimer=2000 #20sec
        carhitcounter=0
        
    # Draw badgers
    if badtimer==0:
        badguys.append([800, random.randint(50,550)])
        
        # badguy appears faster and faster (up to a max)
        badtimer=100-(badtimer1*2)
        if badtimer1>=30:
            badtimer1=30
        else:
            badtimer1+=1
    # -------------- 
    
    # Car move
    # assumption only 1 car, drives only from bottom to top of screen
    for car in cars:
        if car[1]<-255:
            cars.clear()
        if car[1]>200:
            car[1]-= 1 # slow
        elif car[1]==200:
            if carpause>0:
                #wait
                carpause-= 1
            elif carpause==0:
                car[1]-= 1
        else:
            car[1]-= 3 
    
    # Badguys
    index=0
    for badguy in badguys:
        if badguy[0]<-64: # security if gone left too much without detection
            badguys.pop(index)
        if badguy[1]<-64: 
            badguys.pop(index)
        if badguy[1]>600: 
            badguys.pop(index)
        badguy[0]-= random.randint(1,3) # original 7, move to be not constant
        if badtimer1>=30:
            badguy[1]+= random.randint(-4,4)
        
        # Attack castle
        badrect=pygame.Rect(badguyimg.get_rect())
        badrect.top=badguy[1]
        badrect.left=badguy[0]
        if badrect.left<64:
            hit.play()
            healthvalue -= random.randint(5,20)
            badguys.pop(index)
        
        # Badguy shot
        index1=0
        for bullet in arrows:
            bullrect=pygame.Rect(arrow.get_rect())
            bullrect.left=bullet[1]
            bullrect.top=bullet[2]
            if badrect.colliderect(bullrect):
                enemy.play()
                acc[0]+=1
                badguys.pop(index)
                arrows.pop(index1)
                hit_counter+=5
            index1+=1
        
        # Badguy collision with rabbit
        playerrect=pygame.Rect(player.get_rect())
        playerrect.top=playerpos[1]
        playerrect.left=playerpos[0]
        if badrect.colliderect(playerrect):
            scream.play()
            healthvalue -= random.randint(20,40)
            badguys.pop(index)
            hit_counter-=20
            hit_counter=max(hit_counter,0)    

        # Next bad guy
        index+=1
        
    # Carrots
    index2=0
    for med in meds:
        medrect=pygame.Rect(medimg.get_rect())
        medrect.top=med[1]
        medrect.left=med[0]
        
        # Carrots shot
        index1=0
        for bullet in arrows:
            bullrect=pygame.Rect(arrow.get_rect())
            bullrect.left=bullet[1]
            bullrect.top=bullet[2]
            if medrect.colliderect(bullrect):
                enemy.play()
                meds.pop(index2)
                arrows.pop(index1)
            index1+=1
            
        # Carrots eaten
        if medrect.colliderect(playerrect):
            popout.play() 
            healthvalue += 100
            healthvalue=min(healthvalue,194) # cap health
            meds.pop(index2)
            hit_counter+=20 
            
        # Next carrot    
        index2+=1 
    
    # Car (only 1)
    for car in cars:
        carrect=pygame.Rect(carimg.get_rect())
        carrect.top=car[1]
        carrect.left=car[0]
        
        # Car shot
        index1=0
        for bullet in arrows:
            bullrect=pygame.Rect(arrow.get_rect())
            bullrect.left=bullet[1]
            bullrect.top=bullet[2]
            if carrect.colliderect(bullrect):
                arrows.pop(index1)
                carhitcounter+=1
                acc[0]+=1
            if carhitcounter==30:
                trumpets.play()
                cars.clear()
                hit_counter+=50
                meds.append([bullet[1],bullet[2]])
                meds.append([bullet[1]+15,bullet[2]+15])
                meds.append([bullet[1]-5,bullet[2]+30])
            index1+=1
            
        # Car collision
        if carrect.colliderect(playerrect):
            scream.play() 
            healthvalue -= 1
    
    
    # Update
    for badguy in badguys:
        screen.blit(badguyimg, badguy)
    for med in meds:
        screen.blit(medimg, med)
    for car in cars:
        screen.blit(carimg, car)
    
    # Cloud (only 1)
    # Last to be over others on top (diminuate visibility)
    if cloudtimer<=0:
        for cloud in clouds:
            if cloud[0]>800 or cloud[1]>600:
                clouds.clear()
                cloudtimer=random.randint(200,1000)
                mult=1
                directioncounter=0
                clouds.append([random.randint(-640,800), -412])
            else:
                for cloud in clouds:
                    directioncounter+=1
                    if directioncounter>random.randint(50,1000):
                        mult*=(-1)
                        directioncounter=0
                    cloud[0]+=random.randint(0,2)*mult
                    cloud[1]+=random.randint(0,2)
                    screen.blit(cloudimg,(cloud[0],cloud[1])) 
        
        
    # Draw clock
    time_game_mn=int((time_game-(pygame.time.get_ticks()-start_ticks))/60000)    
    time_game_sec=int((time_game-(pygame.time.get_ticks()-start_ticks))/1000%60)
    if (time_game-(pygame.time.get_ticks()-start_ticks))/1000>=10:
        time_color=BLACK
    else:
        time_color=RED
    survivedtext = font_verysmall.render("Time: "+str(time_game_mn)+":"+str(time_game_sec).zfill(2), True, time_color)
    textRect = survivedtext.get_rect()
    textRect.topright=[635,5]
    screen.blit(survivedtext, textRect)

    # Draw health bar
    screen.blit(healthbar, (5,5))
    for health1 in range(healthvalue):
        screen.blit(health, (health1+8,8))
        
    # Draw score
    if hit_counter>=10:
        score_color=BLACK
    else:
        score_color=RED
    scoretext = font_small.render("Score: "+str(hit_counter), True, score_color)
    scoreRect = scoretext.get_rect()
    scoreRect.topright=[400,5]
    screen.blit(scoretext, scoreRect)

    # Update the screen
    pygame.display.flip()
    myClock.tick(100) # ensure frame speed
    
    # loop through the events
    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            running=0
        if event.type == pygame.KEYDOWN:
            if event.key==K_w or  event.key==K_UP :
                keys[0]=True
            elif event.key==K_a or  event.key==K_LEFT:
                keys[1]=True
            elif event.key==K_s or  event.key==K_DOWN:
                keys[2]=True
            elif event.key==K_d or  event.key==K_RIGHT:
                keys[3]=True
            elif event.key==K_ESCAPE:
                running=0
            elif event.key==K_n: # easter egg
                if len(meds)<11: # already pressed
                    meds.append([300,500])
                    meds.append([300,400])
                    meds.append([300,300])
                    meds.append([300,200])
                    meds.append([375,275])
                    meds.append([450,350])
                    meds.append([525,425])
                    meds.append([600,500])
                    meds.append([600,400])
                    meds.append([600,300])
                    meds.append([600,200])
                    trumpets.play()
        if event.type == pygame.KEYUP:
            if event.key==pygame.K_w or  event.key==K_UP:
                keys[0]=False
            elif event.key==pygame.K_a or  event.key==K_LEFT:
                keys[1]=False
            elif event.key==pygame.K_s or  event.key==K_DOWN:
                keys[2]=False
            elif event.key==pygame.K_d or  event.key==K_RIGHT:
                keys[3]=False
        if event.type==pygame.MOUSEBUTTONDOWN:
            if event.button == 1: # left
                if hit_counter>0:
                    shoot.play()
                    position=pygame.mouse.get_pos()
                    acc[1]+=1
                    arrows.append([math.atan2(position[1]-(playerpos1[1]+32),position[0]-(playerpos1[0]+26)),playerpos1[0]+32,playerpos1[1]+32])
                    hit_counter-=1
                else:
                    emptychamber.play()
            if event.button == 3: # right
                if hit_counter>=100:
                    scream6.play()
                    counter=0
                    while index-counter>0:
                        badguys.pop(0) # after a removal, all are shifted, thus nb 0 can be removed again
                        hit_counter-=3    
                        counter+=1
                        if hit_counter<=10: counter=index # forces exit and let some shoots remaining
                    meds.clear()
                else:
                    emptychamber.play()
     # Move player
    if keys[0]:
        playerpos[1]-=5
        playerpos[1]=max(playerpos[1],10)
    elif keys[2]:
        playerpos[1]+=5
        playerpos[1]=min(playerpos[1],590)
    if keys[1]:
        playerpos[0]-=5
        playerpos[0]=max(playerpos[0],10)
    elif keys[3]:
        playerpos[0]+=5
        playerpos[0]=min(playerpos[0],790)
        
    # Win/Lose check
    if (time_game-(pygame.time.get_ticks()-start_ticks))/1000<=0: #original without instructions if pygame.time.get_ticks()>=60000:
        running=0
        exitcode=1
    if healthvalue<=0:
        running=0
        exitcode=0
    if acc[1]!=0:
        accuracy=acc[0]*1.0/acc[1]*100
    else:
        accuracy=0
# -------------- end of loop

# Final display   
pygame.font.init()
if exitcode==0:  
    text = font_verysmall.render("Oups mission failed...", True, RED)
    textRect = text.get_rect()
    textRect.centerx = screen.get_rect().centerx
    textRect.centery = screen.get_rect().centery+24
    screen.blit(gameover, (0,0))
    screen.blit(text, textRect)
else:
    text = font_verysmall.render("Accuracy: "+str(int(accuracy))+"%", True, GREEN)
    textRect = text.get_rect()
    textRect.centerx = screen.get_rect().centerx
    textRect.centery = screen.get_rect().centery+24
    textScore = font_verysmall.render("(final credits: "+str(hit_counter)+")", True, GREEN)
    textScoreRect = text.get_rect()
    textScoreRect.centerx = screen.get_rect().centerx
    textScoreRect.centery = screen.get_rect().centery+48
    screen.blit(youwin, (0,0))
    screen.blit(text, textRect)
    screen.blit(textScore, textScoreRect)

pygame.display.flip()
finalexit=1
while finalexit: # wait for exit
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finalexit=0        
pygame.quit()
sys.exit()
```
